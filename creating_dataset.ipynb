```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
```

```python
datadir="M:\datasets\dogs-cats-images"
categories=["dogs" ,"cats"]
for animal in categories:
    destination=os.path.join(datadir,animal)
    for image in os.listdir(destination):
        img_array=cv2.imread(os.path.join(destination,image),cv2.IMREAD_GRAYSCALE)
        plt.imshow(img_array,cmap="gray")
        plt.show()
        break
    break
```

```python
print(img_array.shape)
```

```python
training_data=[]
img_size=75
def create_trainingdata():
    for animal in categories:
        destination=os.path.join(datadir,animal)
        class_num=categories.index(animal)
        for image in os.listdir(destination):
            try:
                img_array=cv2.imread(os.path.join(destination,image),cv2.IMREAD_GRAYSCALE)
                new_array=cv2.resize(img_array,(img_size,img_size))
                training_data.append([new_array,class_num])
            except Exception as e:
                pass
create_trainingdata()
```

```python
print(len(training_data))
```

```python
import random
random.shuffle(training_data)
#for sample in training_data:
 #   print(sample[1])
```

```python
X=[]
y=[]
```

```python
for features,label in training_data:
    X.append(features)
    y.append(label)
X=np.array(X).reshape(-1,img_size,img_size,1)
```

```python
import pickle
pickle_out=open("X.pickle","wb")
pickle.dump(X,pickle_out)
pickle_out.close()

pickle_out=open("y.pickle","wb")
pickle.dump(y,pickle_out)
pickle_out.close()
```
